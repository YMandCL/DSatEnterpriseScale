# Activation Functions

The Activation Functions described here take as the input the Linear Model, given as

$$z = w_0 + \sum_{i=1}^{N}w_ix_i = \mathbf{x}\mathbf{w},$$

which is described in Chapter 4.  With the exception of the Identity function, the output of the activation is a nonlinear function of the linear model.  The idea behind activation functions is to return a value that can be interpreted as an "on" or "off value when returned to the user as a label.  This interpretation becomes more nuanced in the case of neural networks, but we will return to that later.  The most frequently used activation functions are:

## Identity Function

$$f(z)=z$$

This trivial function recovers the original linear model.

## Logistic Function

$$f(z)=\frac{1}{1+e^{-z}}$$

This function is used in the Logistic Regression algorithm and is the most frequently used activation function in machine learning algorithms.

## Hyperbolic Tangent Function

$$f(z)=\tanh(z)$$

Another very popular function in machine learning, and also deep learning algorithms.

## Rectified Linear Unit (ReLU)

$$f(x)=\max(z,0)$$

This function is popular in deep learning algorithms because of its properties in numerical optimization procedures.

Some initializations and imports here...

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining a z range here
z = np.arange(-1.5,1.5, 0.05)
```

We will define a plotting function to avoid repeating the plot commands to display the functions.

```python
def plotActivation(x, y, titleString, yString):
    plt.plot(x,y)
    plt.xlabel("z")
    plt.ylabel(yString)
    plt.title(titleString)
    plt.grid()
    plt.show()
```

```python
dpi = 300 #dots per square inch when plotting...higher resolution for publications.
plt.rcParams['figure.dpi']= dpi
plt.rcParams["figure.figsize"] = [4,1]

#plotting the 4 functions with adjustments to the 
plotActivation(z,z, 'Identity', 'z')
plotActivation(z,1/(1 + np.exp(-z*5)), 'Logistic', '1/(1+$e^{-5z}$)')
plotActivation(z,np.tanh(z*3),'Hyperbolic Tangent', 'tanh(3z)')
plotActivation(z,np.maximum(z,0),"Rectified Linear Unit (ReLU)", 'max(z,0)')
```
